Enunciado: Escribe una función llamada `scrape_coches_ocasion` que acepta una URL como argumento y devuelve un DataFrame de pandas con información sobre coches de ocasión encontrados en la página web especificada. La función realizará las siguientes tareas:

1. Realizará una solicitud HTTP GET a la [siguiente](https://www.ocasionplus.com/coches-ocasion?type%3DCAR%26sort%3DFECHAOP_DESC%26page%3D2) url.


```python
#https://www.ocasionplus.com/coches-ocasion?v2&orderBy=morePopular
```

```python
url_coches = "https://www.ocasionplus.com/coches-ocasion?v2&orderBy=morePopular"

resultado_coches = requests.get(url_coches)
print(resultado_coches.status_code)
```

2. Comprobará el código de estado de la respuesta. Si el código no es 200 (OK), imprimirá un mensaje de error y retornará `None`.

3. Extraerá los nombres de los coches, los precios rebajados, los precios originales y la información de los kilómetros recorridos de la página.

```python
# creamos el objeto BeautifulSoup para poder acceder al contenido solicitado
sopa_coches = BeautifulSoup(resultado_coches.content, 'html.parser')

# mostramos por pantalla los resultados del objeto de Beautiful Soup. El método ".prettify()" nos ayuda a visualizar los resultados de una forma más amigable
print(sopa_coches.prettify())
```

```python
# sacamos los nombres de los moviles
lista_nombre_producto = sopa_coches.find_all("span", {"class" :"cardVehicle_spot__e6YZx" })
```

```python
lista_nombre_producto1 = sopa_coches.find_all("div", {"class" :"cardVehicle_description__Gp9vd"})
```

```python
#<h1 class="Tittle_brand__9nRGz" data-test="h1-details.brand">KIA Ceed</h1>
#<span class="cardVehicle_spot__e6YZx" data-test="span-brand-model">KIA Ceed</span>
```

```python
print(lista_nombre_producto1)
```

```python
# es el momento de sacar la información útil del método que hemos usado previamente. Para eso lo primero que vamos a hacer es crearnos una lista donde iremos almacenado los resultados que queremos
nombres_productos = []

# dado que es una lista lo que vamos a hacer es iterar por la lista para poder acceder a cada uno de los elementos
for i in lista_nombre_producto1:
    # utilizamos el método ".getText()" para sacar el texto de cada uno de los elementos y lo apendeamos a la lista que hemos creado previamente. 
    nombres_productos.append(i.getText())

print("\n--------------------------\n")

print("Los resultados de extraer el texto de cada uno de los elementos es:\n", nombres_productos)
```
